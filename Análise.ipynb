```python
import pandas as pd
import seaborn as sns
```

```python
relatorio = pd.read_csv('exp1.csv', index_col=False, sep=';')
relatorio
```

```python
# ver o resultado médio por técnica de desbalanceamento
relatorio.groupby(by=['Tec-Desbal']).mean(numeric_only=True)
```

```python
# selecionar o melhor resultado de todos dado uma métrica
idx = relatorio['Acuracia Balanceada'].idxmax()
print(relatorio.loc[idx])
```

```python
# selecionar o top5 resultados dado uma métrica
top5_resultados = relatorio.nlargest(5, 'Acuracia Balanceada')
top5_resultados
```
